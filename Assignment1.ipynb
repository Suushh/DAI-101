```python
#DATAVISUALISATION
#Countplot of categorical Columns
```

```python
for col in df:
    if df[col].dtype == 'O':
        sns.countplot(x=col,data=df,palette='Set2')
        plt.show()
```

```python
num_col = ['Hours_Studied', 'Attendance', 'Previous_Scores']
cat_col = ['Parental_Involvement', 'Access_to_Resources', 'Sleep_Hours', 'Extracurricular_Activities', 'Motivation_Level', 'Internet_Access', 'Tutoring_Sessions', 'Family_Income', 'Teacher_Quality', 'School_Type', 'Peer_Influence', 'Physical_Activity', 'Learning_Disabilities', 'Parental_Education_Level', 'Distance_from_Home', 'Gender']
target = 'Exam_Score'
```

```python
#Histograms of numerical columns
```

```python
for column in num_col:
    plt.figure(figsize=(10, 5))
    sns.histplot(df[column], bins=30)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
df_new=df[df["Exam_Score"]<=100]
```

```python
df_new.shape
```

```python
plt.figure(figsize=(15, 6))
sns.countplot(x=df_new[target], color='lightgreen')

plt.title("The distribution of taget variable 'exam_score'.")
plt.grid(True, which='major', axis='y', linestyle='--', linewidth=0.7)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes = axes.ravel()

for idx, col in enumerate(num_col):
    sns.scatterplot(x=df_new[col], y=df_new[target], ax=axes[idx], color='yellowgreen')
    axes[idx].set_title(f'Relationship b/t {col} & {target}')
    plt.xlabel(col)
    plt.ylabel(target)

plt.tight_layout()
plt.show()
```

```python
corr_col = ['Hours_Studied', 'Attendance', 'Previous_Scores', 'Exam_Score']
df_corr = df_new[corr_col]

corr = df_corr.corr()

plt.figure(figsize=(5,4))
sns.heatmap(corr, annot=True, cmap='Greens')
```

```python
fig, axes = plt.subplots(8, 2, figsize=(15, 60))

axes = axes.ravel()

for idx, col in enumerate(cat_col):
    sns.boxplot(x=df_new[col], y=df_new[target], ax=axes[idx])
    axes[idx].set_title(f'Relationship between {col} and {target}')
    plt.xlabel(col)
    plt.ylabel(target)

plt.tight_layout()
plt.show()
```
